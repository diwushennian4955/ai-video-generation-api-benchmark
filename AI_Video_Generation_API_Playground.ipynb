```python
# Generate a video — select your model and prompt

# ===== CONFIGURE HERE =====
MODEL = 'kling-v3-pro'  # Options: sora-2, veo-3, runway-gen4, kling-v3-pro
PROMPT = 'A cinematic drone shot over a misty mountain range at golden hour, 4K quality'
DURATION = 5  # seconds
ASPECT_RATIO = '16:9'
# ==========================

print(f'Generating video with {MODEL}...')
print(f'Prompt: "{PROMPT}"')
print(f'Duration: {DURATION}s | Aspect: {ASPECT_RATIO}')
print(f'Cost: ~$0.05–$0.20 (vs $0.30–$2.00 direct)\n')

try:
    response = client.video.generate(
        model=MODEL,
        prompt=PROMPT,
        duration=DURATION,
        aspect_ratio=ASPECT_RATIO,
        quality='cinematic'
    )
    print(f'✅ Video URL: {response.video_url}')
    print(f'Generation time: {response.generation_time}s')
    
    # Display video
    from IPython.display import HTML
    display(HTML(f'<video width="640" controls><source src="{response.video_url}" type="video/mp4"></video>'))
    
except Exception as e:
    print(f'Note: Add your API key above to run live demo')
    print(f'Get free key at: https://nexa-api.com')
```

```python
# Cost estimator
import ipywidgets as widgets
from IPython.display import display as ipy_display

videos_slider = widgets.IntSlider(value=100, min=10, max=10000, step=10, description='Videos/month:')
output = widgets.Output()

def update_cost(change):
    with output:
        output.clear_output()
        n = change['new']
        print(f'\n💰 Cost for {n} videos/month:')
        print(f'  NexaAPI:      ${n * 0.10:.2f}–${n * 0.20:.2f}/month')
        print(f'  Sora 2 direct: ${n * 0.50:.2f}–${n * 2.00:.2f}/month')
        print(f'  Veo 3 direct:  ${n * 0.30:.2f}–${n * 1.50:.2f}/month')
        print(f'  Runway direct: ${n * 0.25:.2f}–${n * 1.00:.2f}/month')
        savings = (n * 0.75 - n * 0.15) * 12
        print(f'\n  Annual savings with NexaAPI: ~${savings:,.0f}')
        print(f'  → https://nexa-api.com')

videos_slider.observe(update_cost, names='value')
ipy_display(videos_slider, output)
update_cost({'new': 100})
```
